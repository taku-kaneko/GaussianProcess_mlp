# Multiple Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

## Prepare training data

```python
x1 = np.linspace(-5, 5, 20)
x2 = np.linspace(-5, 5, 20)
X1, X2 = np.meshgrid(x1, x2)
y = 1 + 1.2 * X1 + 0.6 * X2 + np.random.randn(20 * 20).reshape(20, 20)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter3D(np.ravel(X1), np.ravel(X2), np.ravel(y), marker="o", linestyle="None", color="red")
```

## Define OLS function

```python
def ols(y, X):
    w = np.linalg.multi_dot([np.linalg.inv(np.dot(X.T, X)), X.T, y])
    return w
```

```python
X1_ravel = np.ravel(X1)
X2_ravel = np.ravel(X2)
X = np.ones((len(X1_ravel), 3))
X[:, 1] = X1_ravel
X[:, 2] = X2_ravel
w = ols(np.ravel(y), X)
print(w)
```

```python
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
X1_new, X2_new = np.meshgrid(x1, x2)
y_hat = w[0] + w[1] * X1_new + w[2] * X2_new
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X1_new, X2_new, y_hat, color='lightskyblue', linewidth=0.5)
ax.scatter3D(np.ravel(X1), np.ravel(X2), np.ravel(y), marker="o", linestyle="None", color="red")
```
